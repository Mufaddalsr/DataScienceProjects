```
import pandas as pd
pdf = dbframe.toPandas()
pdf

def f(pdf):
  val = 0
  if pdf['Sex'] == 'Male':
    if (13.2 <= pdf['Hb'] <= 16.6) :
      pass
    else:
      val+=1
    if (38.3 <= pdf['Hct'] <= 48.6) :
      pass
    else:
      val+=1
    if (pdf['Systolic'] != 120.0) :
      val+=1
    if (pdf['Diastolic'] != 80.0) :
      val+=1
    if (0.6<= pdf['Creatinine'] <= 1.2) :
      pass
    else:
      val+=1


  if pdf['Sex'] == 'Female':
    if (11.6 <= pdf['Hb'] <= 15.0) :
      pass
    else:
      val+=1
    if (35.5 <= pdf['Hct'] <= 44.9) :
      pass
    else:
      val+=1
    if (pdf['Systolic'] != 120.0) :
      val+=1
    if (pdf['Diastolic'] != 80.0) :
      val+=1
    if (0.5<= pdf['Creatinine'] <= 1.1) :
      pass
    else:
      val+=1

  

  if val >=6:
    verdict = "Patient is Having Severe Uncertainties in Test Results, goo to nutritionist  "
  elif val >=4:
    verdict = "Patient Needs to Consult a nutritionist Soon"
  elif val >=2:
    verdict = "Patient can Consult a nutritionist (Note: Not Urgent)"
  else:
    verdict = "Patient is Totally Fine, No Need to Consult a nutritionist"
  return verdict


pdf['assessment'] = pdf.apply(f, axis=1)
pdf.dtypes
```

```
#sparkDF=spark.createDataFrame(pandasDF.astype(str)) 
from pyspark.sql.types import StructType,StructField, StringType, IntegerType,FloatType
mySchema = StructType([ StructField("Name", StringType(), True)\
                       ,StructField("Age", IntegerType(), True)
                       ,StructField("Sex", StringType(), True)
                       ,StructField("Hb", FloatType(), True)
                       ,StructField("Hct", FloatType(), True)
                       ,StructField("Systolic", IntegerType(), True)
                       ,StructField("Diastolic", IntegerType(), True)
                       ,StructField("Creatinine", FloatType(), True)
                       ,StructField("assessment", StringType(), True)
                       ])

dbframe = spark.createDataFrame(pdf,schema=mySchema)
# dbframe.printSchema()
# dbframe.show()

# Enable Apache Arrow to convert Pandas to PySpark DataFrame
spark.conf.set("spark.sql.execution.arrow.enabled","true")
dbframe=spark.createDataFrame(pdf) 
# dbframe.printSchema()
dbframe.head(100)
```
